```python
# setting input and output with jupyter notebook in context 
if 'ipykernel_launcher.py' in sys.argv[0]:
    input_fn = 'results/samplesheets/post-hicpro/{}.readlength.google-samplesheet.tsv'.format(latest_date)
    output_prefix = 'results/samplesheets/post-hicpro/readlength.samplesheet'
else:
    input_fn = sys.argv[1]
    output_prefix = sys.argv[2]
```

```python
# loading the samplesheet
df = pd.read_table(input_fn, skiprows=0)
```

```python
df
```

```python
# extract those samples which are ready for post-hicpro analyses and processing
ready_df = df.loc[(df['Start Processing'] == 1) & (df['Download Status'] == 1) & (df['HiCPro Status'] == 1)]

# extract only the columns needed
major_cols = ['Sample Name (as used in the server)',
 'GSE ID',
 'Organism',
 'Biological Replicate Serial No',
 'Technical Replicate Serial No',
 'ChIP-seq Pull Down',
 'Restriction Enzyme',
 'Decided Read Length']
ready_df = ready_df[major_cols]
```

```python
ready_df
```

```python
# capitalizes organism
def parse_organism(string):
    new_words = []
    for word in string.split():
        new_words.append(word.capitalize())
    new_string = '_'.join(new_words)
    return(new_string)

ready_df.loc[:, 'Organism'] = ready_df.loc[:, 'Organism'].apply(parse_organism)
```

```python
# getting the sample names
sample_names = []
for i, sr in ready_df.iterrows():
    sample_name = '{sample_name}.{gse_id}.{organism}.{antibody_target}.b{biological_rep}'
    sample_name = sample_name.format(sample_name=sr[0],
                                     gse_id=sr[1],
                                     organism=sr[2],
                                     antibody_target=sr[5], 
                                     biological_rep=sr[3])
    sample_names.append(sample_name)
ready_df.loc[:, 'sample_name'] = sample_names
```

```python
# renaming the columns for easy computational use 
ready_df.columns = ['sample_name', 'gse_id',
                    'organism', 'bio_rep', 'tech_rep', 'antibody_target',
                    'restriction_enzyme', 'read_length', 'std_sample_name']
```

```python
# reorder the columns
reorder = ['std_sample_name',
             'gse_id',
             'organism',
             'bio_rep',
             'tech_rep',
             'antibody_target',
             'restriction_enzyme',
             'read_length',
             'sample_name']
ready_df = ready_df[reorder]
final_df = ready_df.drop_duplicates(subset=['std_sample_name', 'read_length'])
final_df = final_df.drop(['sample_name'], axis=1)
```

```python
final_df
```

```python
# getting the r1 and r2 read lengths
r1_lengths = []
r2_lengths = []
for i, sr in final_df.iterrows():
    if ',' in sr[7]:
        r1 = sr[7].strip().split(',')[0]
        r2 = sr[7].strip().split(',')[1]
    else:
        r1 = sr[7].strip()
        r2 = sr[7].strip()
    r1_lengths.append(r1)
    r2_lengths.append(r2)
final_df.loc[:, 'r1_lengths'] = r1_lengths
final_df.loc[:, 'r2_lengths'] = r2_lengths
```

```python
final_df
```

```python
readlength_cols = ['std_sample_name', 'r1_lengths', 'r2_lengths']
readlength_df = final_df[readlength_cols]
readlength_df = readlength_df.reset_index(drop=True)
```

```python
readlength_df
```

```python
without_header_output = '{}.without_header.tsv'.format(output_prefix)
readlength_df.to_csv(without_header_output, header=False, index=False, sep='\t')
```
